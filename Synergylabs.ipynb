```python
from darkflow.net.build import TFNet
import cv2
import tensorflow as tf


config = tf.ConfigProto(log_device_placement = False)
config.gpu_options.allow_growth = False 

with tf.Session(config=config) as sess:
    options = {
            'model': './cfg/yolo.cfg',
            'load': './yolov2.weights',
            'threshold': 0.6,
               }
    tfnet = TFNet(options)    
```

```python
import cv2

# you put the stitch file in folder 
cap = cv2.VideoCapture('stitch.avi')

while True:
    ret, frame = cap.read()
    
    if ret:
        img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = tfnet.return_predict(img)
        
        for (i, result) in enumerate(results):
            if result['label']=='truck':
                x = result['topleft']['x']
                w = result['bottomright']['x']-result['topleft']['x']
                y = result['topleft']['y']
                h = result['bottomright']['y']-result['topleft']['y']
                cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
                label_position = (x + int(w/8)), abs(y - 10)
                cv2.putText(frame, result['label'] + '  truck_fee_x_to_pay', label_position , cv2.FONT_HERSHEY_SIMPLEX,0.5, (0,0,255), 2)
            if result['label']=='car':
                x = result['topleft']['x']
                
                w = result['bottomright']['x']-result['topleft']['x']
                y = result['topleft']['y']
                h = result['bottomright']['y']-result['topleft']['y']
                cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
                label_position = (x + int(w/8)), abs(y - 10)
                cv2.putText(frame, result['label'] + '  car_fee_p_to_pay', label_position , cv2.FONT_HERSHEY_SIMPLEX,0.5, (0,0,255), 2)
            if result['label']=='bus':
                x = result['topleft']['x']
                w = result['bottomright']['x']-result['topleft']['x']
                y = result['topleft']['y']
                h = result['bottomright']['y']-result['topleft']['y']
                cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
                label_position = (x + int(w/8)), abs(y - 10)
                cv2.putText(frame, result['label'] + '  bus_fee_z_to_pay', label_position , cv2.FONT_HERSHEY_SIMPLEX,0.5, (0,0,255), 2)
            if result['label']=='lorry':
                x = result['topleft']['x']
                w = result['bottomright']['x']-result['topleft']['x']
                y = result['topleft']['y']
                h = result['bottomright']['y']-result['topleft']['y']
                cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
                label_position = (x + int(w/8)), abs(y - 10)
                cv2.putText(frame, result['label'] + '  lorry_fee_y_to_pay', label_position , cv2.FONT_HERSHEY_SIMPLEX,0.5, (0,0,255), 2)

        cv2.imshow("vehicles Detection at toll using YOLO(darkflow)", frame)

        if cv2.waitKey(1) == 13: #13 is the Enter Key
            break

cap.release()
cv2.destroyAllWindows()
```
